```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from tabulate import tabulate
import pandas as pd
import numpy as np
```

```python
def response_await(url):
    response = urlopen(url)
    data = response.read()
    Soup = BeautifulSoup(data, 'lxml')
    return Soup
```

```python
arr = ['2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']

part1 = 'https://www.iplt20.com/stats/'
part2='/best-batting-average'
maximum = 101
data = []
for i in range(0, 12):
    url = part1 + arr[i]+ part2
    Soup = response_await(url)
    data_yr = []
    rows = Soup.find_all('tr')
    count = 1
    for row in rows[1:]:
        name = row.get('class')
        if(count == 1):
            data.append([arr[i], name[2], maximum-count])
        else:
            data.append([arr[i], name[1], maximum-count])
        count += 1
    #data.append(data_yr)
print(data)
```

```python
df = pd.DataFrame(data,columns=['Year', 'Team', 'Points'])
```
